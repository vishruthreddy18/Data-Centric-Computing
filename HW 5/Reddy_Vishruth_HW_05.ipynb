```python
for i in sdf.select_dtypes(exclude=['number']).columns:
    row_count = sdf[i].value_counts()
    print(i,'\n', row_count)
```

```python
unique_count = {}
for i in sdf.select_dtypes(include=['number']).columns:
    unique_count[i] = sdf[i].nunique()

row_count = sdf[min(unique_count)].value_counts()
print(i,'\n', row_count)
```
